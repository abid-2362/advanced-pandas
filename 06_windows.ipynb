```python
import pandas as pd
import os
```

```python
scoring = pd.read_pickle(os.path.join('data', 'modified', 'scoring_mi.pickle'))
birth_dates = pd.read_pickle(os.path.join('data', 'modified', 'birth_dates.pickle'))
scoring.head()
```

```python
scoring.index
```

```python
wayne = scoring.loc['gretzwa01']
wayne
```

```python
sums = wayne.groupby(level=0)['G', 'A'].sum()
sums
```

```python
means = sums.rolling(3)['G'].mean()
assist_means = sums.rolling(3)['A'].mean()
means.head(10)
```

```python
sums.assign(goals_rolling_mean = means, assist_rolling_mean=assist_means).plot()
```

```python
means = sums.rolling(3, min_periods=2)['G'].mean()
means.head(10)
```

```python
means = sums.rolling(3).mean()
means.head()
```

### Expanding windows


```python
sums.expanding().sum().plot()
```

```python
sums.expanding().max().plot()
```

```python
sums.cummax().plot()
```

```python
sums.cumsum().plot()
```

```python
totals = sums.sum()
totals
```

```python
total_part = sums['G'].expanding().aggregate(lambda x:
                                             x.sum() / totals['G'])
total_part.plot()
```
